```python
import pandas as pd

df = pd.read_csv('amg_e53.csv')
df.sort_values(['car_price'])
```

```python
# feature selection\n
# what kind of car's feature do you want?
# features here are strings WTF...

# Task: Data Cleaning

def str_to_list(u):
    if type(u) == list: 
        return u
    u = u.replace('\\', '').replace('\'', '').replace('/', '').lower()
    res = u[1:-2].split(', ')
    return res

def df_feature_to_list(df):
    tmp_df = df
    tmp_df['features_list'] = df.features_list.apply(str_to_list)
    return tmp_df

# Task: get all feature
def all_features(df):
    x = df['features_list']
    res = set()
    for (i, v) in x.items():
        v = v.replace('\\', '').replace('\'', '').replace('/', '').lower()[1:-2].split(', ')
        res = res | set(v)
    return list(res)
                                          
wish_list = ['panorama', 'burmester', 'blind spot', 'driver assistance package',\
             'parking assistance package', 'head-up']

final_df = df_feature_to_list(df)

#assume df.features_list is a LIST
def query_by_wished_feature(df, wlst):
    mask = df.features_list.apply(lambda x: any(item for item in wlst if item in x))
    return df[mask]
nice = query_by_wished_feature(final_df, wish_list).sort_values(['car_type', 'car_price'])
                                          
for i, x in nice['car_details_url'].items():
    print(i, x)
                                          
nice
```
